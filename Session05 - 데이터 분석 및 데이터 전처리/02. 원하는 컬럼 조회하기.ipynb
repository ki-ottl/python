# 원하는 컬럼 조회하기

```python
# 내가 계산을 하기 위해서 필요한 컬럼은 CUSTID랑 AVGPRICE라면,
# 그 컬럼만 조회할 수 있다.
```

```python
# 데이터프레임명.loc[:,[컬럼명]] 
# [row , column] 이 들어간다고 기억!
# :은 전체 행을 전부 다 조회한다는 의미 
# ,를 하면 내가 원하는 컬럼명을 명시해줄 수 있다
```

```python
# 데이터프레임명.iloc[:,[인덱스]]
# 컬럼에도 인덱스가 있어서 인덱스로 접근할 수 있다.
# 이때는 index location이라고 해서 iloc을 사용한다.
```

## 데이터프레임명.loc[: , [컬럼명]] 

```python
# CUSTID와 AVGPRICE 컬럼만 조회하기
```

```python
import pandas as pd
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")
```

```python
customerData
```

```python
# case 1
targetColumns = customerData.loc[: , ["CUSTID", "AVGPRICE"] ]
targetColumns
```

```python
# case 2
# 정리해서 조회
```

```python
targetColumns = ["CUSTID", "AVGPRICE"]
```

```python
columnSelectVal = customerData.loc[: , targetColumns ]
columnSelectVal
```

```python
# case 3
# 간단하게 조회
customerData [ targetColumns ]
```

## 데이터프레임명.iloc[:,[인덱스]]

```python
customerData = pd.read_csv("../dataset/customerdata.csv")
customerData
```

```python
# 전체 행과 전체 컬럼을 모두 조회
customerData.iloc[ : , : ]
```

```python
# 전체 행을 조회할 것이고, 1 이후의 컬럼만 조회할 것이다.
# AVGPRICE가 1이라는 인덱스를 갖고 있기 때문에 AVGPRICE부터 조회
customerData.iloc [ :, 1:]
```

```python
# 인덱스로도 조회가 가능하다
avgColumn = 1
emiColumn = 2
```

```python
customerData.iloc [ :, [avgColumn,emiColumn] ]
```

### 해당하는 컬럼의 인덱스를 찾는 방법
#### customerData.columns.get_loc("컬럼명")

```python
# 코드의 가독성은 떨어지지만, 재사용성이 좋다.
```

```python
customerDataColumns = customerData.columns
```

```python
avgColumn = customerDataColumns.get_loc("AVGPRICE")
emiColumn = customerDataColumns.get_loc("EMI")
```

```python
customerData.iloc[ : , [avgColumn, emiColumn]]
```
