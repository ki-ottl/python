```python
# 자주 사용
# np.where(조건식, 참일 경우 반환, 거짓일 경우 반환)
# 엑셀로 상한선, 하한선 조건식 썼던 것
np.where(customerData.PRODUCTAGE < 1 , 1, 0)  # <1 인 경우에는 1 그 외에는 기존값
```

```python
# 데이터를 조작하는 방법

# 다중으로 할 경우!
# np.where 안에다가 np.where로 계속 이어주기!

# 엑셀에서 했던 상한선, 하한선 처리를 기억하자!
# if (셀값 > 상한값 , 상한값, if(셀값<하한선), 하한값)
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")
```

```python
np.where(customerData.PRODUCTAGE < 1 , 1,  
         np.where (customerData.PRODUCTAGE < 2, 2, 
                   np.where(customerData.PRODUCTAGE < 3, 3, 4)))
```

```python
customerData["PRODUCT_AGE_NEW"] = np.where(customerData.PRODUCTAGE < 1 , 1,
                                           np.where (customerData.PRODUCTAGE < 2, 2,
                                                     np.where(customerData.PRODUCTAGE < 3, 3, 4)))
```

## 데이터 조작 case2. 함수 사용

```python
# 데이터를 조작하는 두번째 방법
# np.where는 빠르다. 하지만 복잡한 로직을 적용하기에는 힘들다.
# 복잡한 로직인 경우에는 함수를 이용해서 조작하는 것이 적절하다.
```

```python
inValue = 1.5
outValue = 1
```

```python
if inValue < 1 :
    outValue = 1
elif inValue < 2 :
    outValue = 2
elif inValue < 3 :
    outValue = 3
else :
    outValue = 4
```

```python
def productAgeDefine(inValue) :
    
    # inValue = 1.5
    outValue = 1
    
    if inValue < 1 :
        outValue = 1
    elif inValue < 2 :
        outValue = 2
    elif inValue < 3 :
        outValue = 3
    else :
        outValue = 4
        
    return outValue
```

```python
# 함수를 사용해서 조회를 한 상황
customerData.PRODUCTAGE.apply(productAgeDefine)
```

```python
# 내용을 새로운 컬럼에다가 넣어주겠다
customerData["PRODUCTAGE_NEWF"] = customerData.PRODUCTAGE.apply(productAgeDefine)
```
