# 06. 데이터 정렬하기

```python
# customerData.sort_values(by = [정렬컬럼], -> 정렬키 설정
#                          ascending = [정렬기준], -> 오름차순 or  내림차순
#                          ignore_index = False or True, -> True 선택 시, 인덱스 재정렬
#                          inplace = False or True) -> True 선택 시, 데이터프레임에 정렬된 내용을 바로 반영
```

```python
import pandas as pd
```

```python
customerData = pd.read_csv("../dataset/customerData.csv")
```

```python
customerData.head()
```

```python
# 정렬키 선정
sortKey = ["AVGPRICE"]
```

```python
customerData.sort_values(by = sortKey,
                         ascending = True,
                         ignore_index = False) 
```

```python
# ignore_index = True로 인덱스 재정렬
customerData.sort_values(by = sortKey,
                         ascending = True,
                         ignore_index = True) 
```

```python
sortedData = customerData.sort_values(by = sortKey,
                                      ascending = True,
                                      ignore_index = True) 
sortedData
```
