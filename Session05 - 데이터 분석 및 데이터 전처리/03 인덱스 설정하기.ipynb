# 03 인덱스 설정하기

### reset_index() : 인덱스를 재생성

### set_index() : 특정 컬럼을 인덱스로 설정

```python
# 인덱스를 재생성하고, 기존에 있었던 인덱스는 컬럼으로 남기겠다. ex) emi3Cs.reset_index()
# 기존의 인덱스를 삭제하고 싶다면 drop = True 사용 ex) emi3Cs.reset_index(drop = True)
# 특정 컬럼을 인덱스로 설정해줄 수 있다 ex) emi3Cs.set_index( keys = ["CUSTID"] )
# 중요한 것은 조회한 결과를 변수에 담아줘야 한다는 점 유의
```

```python
# 변수에 다시 담지 않고, 바로 적용을 시켜주는 parameter인 inplace : emi3Cs.reset_index(drop = True, inplace = True)
```

```python
import pandas as pd
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")
```

```python
customerData
```

```python
emi3Cs = customerData.loc [ customerData.EMI == 3 ] 
```

```python
# EMI가 3인 것들만 뽑아서 조회를 하다보니, 인덱스를 재정렬 할 필요가 있다.
emi3Cs
```

```python
# 인덱스를 재생성하고, 기존에 있었던 인덱스는 컬럼으로 남기겠다.
# 중요한 것은 조회한 결과를 변수에 담아줘야 한다는 점
emi3CsIndex = emi3Cs.reset_index()
emi3CsIndex
```

```python
# 기존의 인덱스를 삭제하고 싶다면 drop = True 사용
# 중요한 것은 조회한 결과를 변수에 담아줘야 한다는 점
emi3CsIndexNew = emi3Cs.reset_index(drop=True)
emi3CsIndexNew
```

```python
# 변수에 다시 담지 않고, 바로 적용을 시켜주는 옵션
emi3Cs.reset_index(drop = True, inplace = True)
```

```python
# reset된 결과가 그대로 남아있는 것을 알 수 있다
emi3Cs
```

```python
# 특정 컬럼을 인덱스로 설정해줄 수 있다
emi3CsIndexSet = emi3Cs.set_index( keys = ["CUSTID"] )
emi3CsIndexSet
```

```python
# 인덱스이기 때문에 아래와 같이 조회하는 것도 가능하다
emi3CsIndexSet.loc ["A13566" : "I12899" ]
```
